```python
import pandas as pd
import numpy as np
import geopandas as gpd
import os
```

```python
# Read in destination files as geodataframes
path = './unsynced-data/rwanda/destinations'

# list the files in the path directory and if they end in .parquet, add them to a list called files
files = [file for file in os.listdir(path) if file.endswith('.parquet')]
print (files)
health_posts = gpd.read_parquet(path + '/' + files[0])
dense_urban = gpd.read_parquet(path + '/' + files[1])
primary_schools = gpd.read_parquet(path + '/' + files[2])
bridge_sites = gpd.read_parquet(path + '/' + files[3])
health_centers = gpd.read_parquet(path + '/' + files[4])
urban_center = gpd.read_parquet(path + '/' + files[5])
major_hospitals = gpd.read_parquet(path + '/' + files[6])
semi_dense_urban = gpd.read_parquet(path + '/' + files[7])
all_education_facilities = gpd.read_parquet(path + '/' + files[8])
secondary_schools = gpd.read_parquet(path + '/' + files[9])
suburban = gpd.read_parquet(path + '/' + files[10])
all_health_facilities = gpd.read_parquet(path + '/' + files[11])
datasets = [health_posts, dense_urban, primary_schools, bridge_sites, health_centers, urban_center, major_hospitals, semi_dense_urban, all_education_facilities, secondary_schools, suburban, all_health_facilities]
for dataset in datasets:
    dataset.crs = 'epsg:4326'
```

```python
# change bridge_sites['subregion_indices'] to string
bridge_sites['subregion_indices'] = bridge_sites['subregion_indices'].astype(str)
# remove the brackets from the subregion_indices column
bridge_sites['subregion_indices'] = bridge_sites['subregion_indices'].str.replace('[', '')
bridge_sites['subregion_indices'] = bridge_sites['subregion_indices'].str.replace(']', '')
# replace the " " with _
bridge_sites['subregion_indices'] = bridge_sites['subregion_indices'].str.replace(' ', '_')
bridge_sites
```

```python
# write each dataset to a geojson file
num = 0
geojson_files = []
for dataset in datasets:
    filename = path + '/' + files[num][:-8] + '.geojson'
    geojson_files.append(filename)
    print(filename)
    dataset.to_file(filename, driver='GeoJSON')
    num += 1
```

```python
for f in geojson_files:
    print(f)
```
